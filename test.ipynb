```python
import numpy as np
# 1D array
a= np.array([2,4,3,5,6])
b= np.array([[2,3,5,5,4],
             [4,2,1,23,5]])
c= np.array([
    [
        [6,7,7],
        [5,4,5]
    ],
    [
        [0,8,3],
        [2,45,22]
    ],
    [
        [65,64,0],
        [44,21,17]
     ]
     ])

print(f"This is a 1D array:\n{a}\n")
print(f"This is a 2D array:\n{b}\n")
print(f"This is a 3D array:\n{c}")
```

```python
new_array= np.arange(0,2000,2)
print(f"The new array is {new_array}")
sum=new_array.sum()
# size=new_array.size()
mean= new_array.mean()
min=new_array.min()
print(f"The sum of the array is: {sum}")
# print(f"The size of the array is: {size}")
print(f"The mean of the array is: {mean}")
print(f"The min of the array is: {min}")
```
