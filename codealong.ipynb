```python
# importing libries
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# create row vector
row_vector = np.array([2,5,7,8])
print(f"This is the row vector: \n{row_vector}")
```

```python
# column vector (2D with shape 3x1)
col_vector = np.array([[21],[14],[10],[11]])
print(f"This is the column vector: \n{col_vector}")
col_vector[0]
col_vector[3]
```

### Vector addition

```python
# addition of 2 vectors together
vec_a = np.array([8,9])
vec_b = np.array([4,8])
result= vec_a + vec_b
print (f"The addition of the 2 vectors is: \n {result}")
```

```python
# adding 2 column vector together
col_vector1 = np.array([[21],[14],[10],[11]])
col_vector2 = np.array([[2],[17],[1],[17]])
col_add= col_vector1 + col_vector2
print (f"The addition of the 2 column vectors is: \n {col_add}")
```

### Geometric interpretation in 2D

**For the addition of 2 vectors above**

```python
plt.figure(figsize=(10,6))
plt.quiver(0,0, vec_a[0], vec_a[1], angles="xy", scale_units="xy", scale=1, color="brown", width=0.005)
plt.quiver(0,0, vec_b[0], vec_b[1], angles="xy", scale_units="xy", scale=1, color="skyblue",width=0.005)
plt.quiver(0,0, result[0], result[1], angles="xy", scale_units="xy", scale=1, color="green",width=0.005)
plt.quiver(vec_a[0],vec_a[1],vec_b[0],vec_b[1],angles="xy", scale_units="xy",scale=1,color="red",width=0.005)
plt.xlim(-1, 15)
plt.ylim(-1, 20)
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule")
plt.legend(['Vector a','Vector b', 'a + b'])
plt.show()
```

### Properties of vector addition

**Commutative property:**

```python
# it says A+B = B+A
a= np.array([3,5,7])
b= np.array([5,6,2])
a_plus_b = a + b
b_plus_a = b + a
print(f"The result for A+B is: {a_plus_b}")
print(f"The result for B+A is: {a_plus_b}")
```

**Associative property:**

```python
# it says A+(B+C) = (A+B)+C 
c= np.array([8,3,1])
left = a + (b + c)
right = (a + b) + c
print(f"The result for A+(B+C) is: {left}")
print(f"The result for (A+B)+C is: {right}")
```

### Scalar Multiplication

**Scalar Multiplication**
 - Multiply every value in a vector by the same number.

 - like upgrading student score across board
 - coverting currency of a give price.

```python
# like upgrading student score across board
student_scores= np.array([12,17,13,10,9,6,20,15,16,20])
upgrade_mark=2
upgraded_student_scores = (student_scores * upgrade_mark)
print(f"The actual students scores are: \n {upgraded_student_scores}")
print(f"The new students scores are: \n {student_scores}")
```

```python
original2d = student_scores
scales=[0.5,1,2,-1,-0.5]
plt.figure(figsize=(10,8))
colors=["red", "green", "purple","blue","black"]
for i, scale in enumerate(scales):
    scaled= scale * original2d
    plt.quiver(0,0,scaled[0],scaled[1],angles="xy", scale_units="xy", scale=1, color=colors[i], width=0.005, label= f"{scale}* v" )
plt.xlim(-20, 30)
plt.ylim(-20, 50)
plt.grid(True)
plt.title("Scalar Multiplication Effects")
plt.legend(title="Used scalars")
plt.show()
```

### Properties of scalar multiplication

```python
a= np.array([7,8,9,3])
c1, c2= 4,2
```

```python
#### Associative: c1(c2*a) = (c1*c2)*a
left= c1*(c2*a)
right= (c1*c2)*a
print (f"c1(c2*a)= {left}")
print (f"(c1*c2)*a = {right}")
print(f'{"Yes, it is" if np.array_equal(left,right) else "No it is not"} Associative')
```

#### Distributive over vector addition: c(a + b) = ca + cb

```python
arr= np.array([7,8,9,3])
c1,b=4,5
left = c1 * (arr + b)
right = c1 * arr + c1 * b
print (f"c(a + b)= {left}")
print (f"ca + cb = {right}")
print(f'{"Yes" if np.array_equal(left,right) else "No"}, it is Distributive')
```

### The Physicist's Perspective of a vector**
    - Magnitude and Direction
    Vectors have both magnitude (length) and direction in space.

```python
# Physicist's view: vectors as arrows in spac
vector_a =np.array([5,3])
magnitude= np.linalg.norm(vector_a)
print(f"Vector is ")
# Direction (unit vector)
unit_vectors= vector_a/magnitude
colors=['red', 'green']
# Visualization
plt.figure(figsize=(10,8))
plt.quiver(0,0,vector_a[0],vector_a[1], angles='xy', scale_units='xy',scale=1, color=colors[0], width=0.005)
plt.quiver(0,0,unit_vectors[0],unit_vectors[1], angles='xy', scale_units='xy',scale=1, color=colors[1], width=0.005)
plt.xlim(-1, 6)
plt.ylim(-1, 5)
plt.title("Magnitude and Direction")
plt.grid(True)
plt.legend(["Original vector","Unit vector" ])
plt.show()
```

### The Mathematician's Perspective**
 - Abstract Objects
 - Vectors are any objects that support addition and scalar multiplication.

```python
# Functions as vectors (function space)
def func1 (x):
    return x**2
def func2 (x):
    return 2*x

def sumfunc(x):
    return func1(x)+ func2(x) #(f+g)(x)= f(x)+ g(x)
# generating numbers
x_vals= np.linspace(0,3,100)
plt.figure(figsize=(10,8))
plt.plot(x_vals, func1(x_vals), label="f(x)=x²")
plt.plot(x_vals, func2(x_vals),label="g(x)= 2x")
plt.plot(x_vals, sumfunc(x_vals),label="(f+g)(x)= f(x) + g(x)")
plt.title("Functions as vectors" )
plt.grid(True)
plt.legend()
plt.show()
```
